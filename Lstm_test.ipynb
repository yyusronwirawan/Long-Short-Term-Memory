```python
# LSTM untuk Prediksi Suhu dan Kelembaban Udara
# Program ini dapat dijalankan di Google Colab

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import warnings
warnings.filterwarnings('ignore')

# Fungsi untuk upload file CSV di Google Colab
def upload_csv_file():
    """
    Fungsi untuk upload file CSV di Google Colab
    """
    from google.colab import files
    print("Silakan upload file CSV Anda:")
    uploaded = files.upload()
    filename = list(uploaded.keys())[0]
    return filename

# Fungsi untuk preprocessing data
def preprocess_data(df, target_columns=['suhu', 'kelembapan'], sequence_length=10):
    """
    Preprocessing data untuk LSTM
    Args:
        df: DataFrame input
        target_columns: kolom target untuk prediksi
        sequence_length: panjang sequence untuk LSTM
    """
    # Konversi kolom waktu ke datetime
    df['waktu'] = pd.to_datetime(df['waktu'])
    df = df.sort_values('waktu').reset_index(drop=True)

    # Pilih kolom numerik yang diperlukan
    feature_columns = ['suhu', 'kelembapan', 'cahaya', 'ktanah']
    data = df[feature_columns].copy()

    # Normalisasi data
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(data)

    # Buat sequences untuk LSTM
    X, y = [], []
    for i in range(sequence_length, len(scaled_data)):
        X.append(scaled_data[i-sequence_length:i])
        y.append(scaled_data[i, [feature_columns.index(col) for col in target_columns]])

    X = np.array(X)
    y = np.array(y)

    return X, y, scaler, feature_columns

# Fungsi untuk membuat dan melatih model LSTM
def create_lstm_model(input_shape, output_dim=2):
    """
    Membuat model LSTM
    Args:
        input_shape: bentuk input (sequence_length, n_features)
        output_dim: jumlah output (2 untuk suhu dan kelembaban)
    """
    model = Sequential([
        LSTM(50, return_sequences=True, input_shape=input_shape),
        Dropout(0.2),
        LSTM(50, return_sequences=True),
        Dropout(0.2),
        LSTM(50),
        Dropout(0.2),
        Dense(25),
        Dense(output_dim)
    ])

    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='mse',
        metrics=['mae']
    )

    return model

# Fungsi untuk membuat prediksi masa depan
def predict_future(model, last_sequence, scaler, n_future=10, target_indices=[0, 1]):
    """
    Prediksi nilai masa depan
    Args:
        model: model LSTM yang sudah dilatih
        last_sequence: sequence terakhir dari data
        scaler: scaler yang digunakan untuk normalisasi
        n_future: jumlah prediksi ke depan
        target_indices: indeks kolom target dalam scaler
    """
    predictions = []
    current_sequence = last_sequence.copy()

    for _ in range(n_future):
        # Prediksi step selanjutnya
        pred = model.predict(current_sequence.reshape(1, *current_sequence.shape), verbose=0)
        predictions.append(pred[0])

        # Update sequence untuk prediksi selanjutnya
        new_row = current_sequence[-1].copy()
        new_row[target_indices] = pred[0]
        current_sequence = np.vstack([current_sequence[1:], new_row])

    # Denormalisasi prediksi
    predictions = np.array(predictions)
    dummy_array = np.zeros((len(predictions), scaler.n_features_in_))
    dummy_array[:, target_indices] = predictions
    denorm_predictions = scaler.inverse_transform(dummy_array)

    return denorm_predictions[:, target_indices]

# Fungsi untuk visualisasi hasil
def plot_results(df, predictions, target_columns=['suhu', 'kelembapan'], n_show=50):
    """
    Visualisasi hasil prediksi
    """
    fig, axes = plt.subplots(2, 1, figsize=(12, 10))

    for i, col in enumerate(target_columns):
        # Data aktual (n_show data terakhir)
        actual_data = df[col].tail(n_show).values
        actual_index = range(len(actual_data))

        # Data prediksi
        pred_index = range(len(actual_data), len(actual_data) + len(predictions))

        axes[i].plot(actual_index, actual_data, label=f'{col.title()} Aktual',
                    color='blue', linewidth=2)
        axes[i].plot(pred_index, predictions[:, i], label=f'{col.title()} Prediksi',
                    color='red', linewidth=2, linestyle='--')

        axes[i].axvline(x=len(actual_data)-1, color='gray', linestyle=':', alpha=0.7)
        axes[i].set_title(f'Prediksi {col.title()}')
        axes[i].set_xlabel('Time Steps')
        axes[i].set_ylabel(col.title())
        axes[i].legend()
        axes[i].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

# Main program
def main():
    print("=== PROGRAM PREDIKSI SUHU DAN KELEMBABAN DENGAN LSTM ===\n")

    # 1. Upload dan load data
    print("1. Upload file CSV...")
    filename = upload_csv_file()

    # Load data
    df = pd.read_csv(filename)
    print(f"\nData berhasil dimuat! Shape: {df.shape}")
    print("\nInfo data:")
    print(df.info())
    print(f"\nSample data:\n{df.head()}")

    # 2. Preprocessing
    print("\n2. Preprocessing data...")
    sequence_length = 10  # Gunakan 10 time steps sebelumnya untuk prediksi
    target_columns = ['suhu', 'kelembapan']

    X, y, scaler, feature_columns = preprocess_data(df, target_columns, sequence_length)

    print(f"Shape X: {X.shape}")
    print(f"Shape y: {y.shape}")

    # 3. Split data
    train_size = int(0.8 * len(X))
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    print(f"\nData training: {X_train.shape[0]} samples")
    print(f"Data testing: {X_test.shape[0]} samples")

    # 4. Buat dan latih model
    print("\n3. Membuat dan melatih model LSTM...")
    model = create_lstm_model(input_shape=(X_train.shape[1], X_train.shape[2]))

    print("Arsitektur model:")
    model.summary()

    # Training
    history = model.fit(
        X_train, y_train,
        epochs=50,
        batch_size=16,
        validation_data=(X_test, y_test),
        verbose=1
    )

    # 5. Evaluasi model
    print("\n4. Evaluasi model...")
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)

    # Denormalisasi untuk evaluasi
    dummy_train = np.zeros((len(train_pred), len(feature_columns)))
    dummy_test = np.zeros((len(test_pred), len(feature_columns)))
    dummy_y_train = np.zeros((len(y_train), len(feature_columns)))
    dummy_y_test = np.zeros((len(y_test), len(feature_columns)))

    target_indices = [feature_columns.index(col) for col in target_columns]
    dummy_train[:, target_indices] = train_pred
    dummy_test[:, target_indices] = test_pred
    dummy_y_train[:, target_indices] = y_train
    dummy_y_test[:, target_indices] = y_test

    train_pred_denorm = scaler.inverse_transform(dummy_train)[:, target_indices]
    test_pred_denorm = scaler.inverse_transform(dummy_test)[:, target_indices]
    y_train_denorm = scaler.inverse_transform(dummy_y_train)[:, target_indices]
    y_test_denorm = scaler.inverse_transform(dummy_y_test)[:, target_indices]

    # Hitung metrik
    for i, col in enumerate(target_columns):
        train_rmse = np.sqrt(mean_squared_error(y_train_denorm[:, i], train_pred_denorm[:, i]))
        test_rmse = np.sqrt(mean_squared_error(y_test_denorm[:, i], test_pred_denorm[:, i]))
        train_mae = mean_absolute_error(y_train_denorm[:, i], train_pred_denorm[:, i])
        test_mae = mean_absolute_error(y_test_denorm[:, i], test_pred_denorm[:, i])

        print(f"\n{col.title()}:")
        print(f"  Train RMSE: {train_rmse:.4f}")
        print(f"  Test RMSE: {test_rmse:.4f}")
        print(f"  Train MAE: {train_mae:.4f}")
        print(f"  Test MAE: {test_mae:.4f}")

    # 6. Plot training history
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

    ax1.plot(history.history['loss'], label='Training Loss')
    ax1.plot(history.history['val_loss'], label='Validation Loss')
    ax1.set_title('Model Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    ax2.plot(history.history['mae'], label='Training MAE')
    ax2.plot(history.history['val_mae'], label='Validation MAE')
    ax2.set_title('Model MAE')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('MAE')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # 7. Prediksi masa depan
    print("\n5. Membuat prediksi masa depan...")
    n_future = int(input("Masukkan jumlah prediksi ke depan (default 10): ") or "10")

    # Ambil sequence terakhir
    last_sequence = X[-1]

    # Buat prediksi
    future_predictions = predict_future(
        model, last_sequence, scaler, n_future, target_indices
    )

    print(f"\nPrediksi {n_future} step ke depan:")
    for i, col in enumerate(target_columns):
        print(f"\n{col.title()}:")
        for j, pred in enumerate(future_predictions[:, i]):
            print(f"  Step {j+1}: {pred:.2f}")

    # 8. Visualisasi
    print("\n6. Visualisasi hasil...")
    plot_results(df, future_predictions, target_columns)

    # 9. Analisis statistik data asli
    print("\n7. Analisis statistik data:")
    print(df[target_columns].describe())

    # 10. Export hasil prediksi
    results_df = pd.DataFrame(
        future_predictions,
        columns=[f'prediksi_{col}' for col in target_columns]
    )
    results_df.index = range(1, len(results_df) + 1)
    results_df.index.name = 'step_ke_depan'

    print(f"\nHasil prediksi:")
    print(results_df)

    # Simpan hasil
    results_df.to_csv('hasil_prediksi.csv')
    print("\nHasil prediksi disimpan sebagai 'hasil_prediksi.csv'")

    return model, scaler, feature_columns, target_columns

# Jalankan program utama
if __name__ == "__main__":
    model, scaler, feature_columns, target_columns = main()
```
